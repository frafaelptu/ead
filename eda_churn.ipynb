## Análise Bivariada

```python
# Hipótese: Clientes com Contrato do tipo mensal são mais propensos ao Churn - Quantidade
# Construir uma tabela de Contigência entre Tipo de Contrato e Churn
pd.crosstab(df_churn.Churn, df_churn.Contract, margins=True, margins_name='Total')
```

```python
# Construir uma tabela de Contigência entre Tipo de Contrato e Churn - Proporção
pd.crosstab(df_churn.Churn, df_churn.Contract, margins=True,normalize='index', margins_name='Total')
```

Considerando o publico total, os clientes de contrato mensal representam 55%. Porém, se considerarmos apenas os clientes que abondoram o serviço 88% tinham contrato mensal, desta forma podemos afirmar que há uma correlação o tipo de contrato mensal e o abandono.

```python
# Avaliar a correlação entre duas variáveis categóricas ( qualitativa )

# Executar um teste de hipótese chamado Chi-Square ou Qui-Quadrado de Pearson
#Num teste de hipótese, duas hipóteses são formuladas:
#H0 ( Hipótese Nula): as duas varáveis são independentes
#H1 (Hipótese Complementar): as duas variáveis NÃO são são independentes

# O teste serve para confirmar ou recusar a hipótese nula (H0)
# Quando a probablidade de observarmos H0 é inferior a 0.05 (p-value),
# recusamos a hipótese nula e seguimos com a complementar
```

```python
# Gerar um DF da Crosstab ( sem totais )
df_crosstab_churn_contract = pd.crosstab(df_contracts.Churn, df_churn.Contract)
```

```python
# Cacular os Scores e P-Values
#NumPy e CyPy são dependencias que o pandas já usa por baixo
from scipy.stats import chi2_contingency
#import scipy
```

```python
# Tupla = Lista imutável
chi_scores_churn_contract = chi2_contingency(df_crosstab_churn_contract)
```

```python
chi_scores_churn_contract
```

```python
score_churn_contract = pd.Series(chi_scores_churn_contract[0])
```

```python
pvalues_churn_contract = pd.Series(chi_scores_churn_contract[1])
```

```python
# Apresentar numeros com decimais sem a notação cientifica
pd.set_option('display.float_format', lambda x: '%.15f' % x)
```

```python
df_chi_scores_churn_contract = pd.DataFrame({'Qui2': score_churn_contract, 'p-value': pvalues_churn_contract})
```

```python
# Analisar Scores e P-Value
df_chi_scores_churn_contract
```

Conforme teste estatístico (Chi-Square), o P-Value <= 0.05, desta forma rejeitamos a hipótese nula, ou seja, as variáveis não independentes. Pelo Qui2 alto, podemos afirmar que há uma forte correlação.

```python
# Avaliar a correlação entre uma váriavel qualitativa e quantitativa 
# Usar o mesmo teste de Chi-Square, mas transformar a variável quantitativas em qualitativa
#Hipótese: Cliente com menos de 6 meses de contrato é mais propenso ao Churn

# Criar uma nova coluna no DataFrame com base numa condição de outra coluna
# Coluna chamada 'TempoMenor6Meses'ArithmeticError

import numpy as np
```

```python
# Criar coluna nova
df_churn['TempoMenor6Meses'] = np.where(df_churn.tenure<6,'Yes', 'No')
```

```python
# Checar o DataFrame
df_churn.head(5)
```

```python
df_crosstab_churn_tenure = pd.crosstab(df_churn.Churn, df_churn.TempoMenor6Meses)
```

```python
df_crosstab_churn_tenure
```

```python
chi_scores_churn_tenure = chi2_contingency(df_crosstab_churn_tenure)
```

```python
score_churn_tenure = pd.Series(chi_scores_churn_tenure[0])
pvalues_churn_tenure = pd.Series(chi_scores_churn_tenure[1])
```

```python
df_chi_scores_churn_tenure = pd.DataFrame({'Qui2': score_churn_tenure, 'p-value': pvalues_churn_tenure})
```

```python
df_chi_scores_churn_tenure
```

Como o pvalue <= 0.05, rejeitamos a H0, ou seja, aas variáveis não são indendependentes. E vale mencionar que com base do Qui2 resultante das 2 análises, a correlação entre Churn e Tempod de Contrato < 6 meses é menor que a correlação entre Churn e Tipo de Contrato.

```python
# Correlação entre 2 varáveis numéricas
# Tenure com TotalCharges
# A intuição é que quanto mais tempo de contrato maior o valor pago.

# Correlação entre 2 variáveis numéricas - Pearson

df_churn.tenure.corr(df_churn.TotalCharges)
```

```python
# Correlação entre 2 variáveis numéricas - Spearman

df_churn.tenure.corr(df_churn.TotalCharges, method='spearman')
```

Há uma forte correlação entre Tenure e TotalCharges usando métodos estatísticos de correlação.

```python
df_churn_month
```

```python
# Método de Tukey - IQR ( Distribuições não normal )
# IRQ - Range Interquartil
# IQR = 3 quartil - 1 quartil
q1_TotalCharges_month = df_churn_month.TotalCharges.quantile(0.25)
q3_TotalCharges_month = df_churn_month.TotalCharges.quantile(0.75)
iqr_TotalCharges_month = q3_TotalCharges_month - q1_TotalCharges_month
iqr_TotalCharges_month
```

```python
# Limites inferior e superior
limInferior_TotalCharges_month = q1_TotalCharges_month - (iqr_TotalCharges_month * 1.5)
limSuperiorInf_TotalCharges_month = q3_TotalCharges_month + (iqr_TotalCharges_month * 1.5)
limInferior_TotalCharges_month, limSuperiorInf_TotalCharges_month
```

```python
df_churn_month[(df_churn_month.TotalCharges < limInferior_TotalCharges_month) | (df_churn_month.TotalCharges > limSuperiorInf_TotalCharges_month)]
```

```python
from scipy.stats import zscore
```
