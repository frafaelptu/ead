## AnÃ¡lise Bivariada

```python
# HipÃ³tese: Clientes com Contrato do tipo mensal sÃ£o mais propensos ao Churn - Quantidade
# Construir uma tabela de ContigÃªncia entre Tipo de Contrato e Churn
pd.crosstab(df_churn.Churn, df_churn.Contract, margins=True, margins_name='Total')
```

```python
# Construir uma tabela de ContigÃªncia entre Tipo de Contrato e Churn - ProporÃ§Ã£o
pd.crosstab(df_churn.Churn, df_churn.Contract, margins=True,normalize='index', margins_name='Total')
```

Considerando o publico total, os clientes de contrato mensal representam 55%. PorÃ©m, se considerarmos apenas os clientes que abondoram o serviÃ§o 88% tinham contrato mensal, desta forma podemos afirmar que hÃ¡ uma correlaÃ§Ã£o o tipo de contrato mensal e o abandono.

```python
# Avaliar a correlaÃ§Ã£o entre duas variÃ¡veis categÃ³ricas ( qualitativa )

# Executar um teste de hipÃ³tese chamado Chi-Square ou Qui-Quadrado de Pearson
#Num teste de hipÃ³tese, duas hipÃ³teses sÃ£o formuladas:
#H0 ( HipÃ³tese Nula): as duas varÃ¡veis sÃ£o independentes
#H1 (HipÃ³tese Complementar): as duas variÃ¡veis NÃO sÃ£o sÃ£o independentes

# O teste serve para confirmar ou recusar a hipÃ³tese nula (H0)
# Quando a probablidade de observarmos H0 Ã© inferior a 0.05 (p-value),
# recusamos a hipÃ³tese nula e seguimos com a complementar
```

```python
# Gerar um DF da Crosstab ( sem totais )
df_crosstab_churn_contract = pd.crosstab(df_contracts.Churn, df_churn.Contract)
```

```python
# Cacular os Scores e P-Values
#NumPy e CyPy sÃ£o dependencias que o pandas jÃ¡ usa por baixo
from scipy.stats import chi2_contingency
#import scipy
```

```python
# Tupla = Lista imutÃ¡vel
chi_scores_churn_contract = chi2_contingency(df_crosstab_churn_contract)
```

```python
chi_scores_churn_contract
```

```python
score_churn_contract = pd.Series(chi_scores_churn_contract[0])
```

```python
pvalues_churn_contract = pd.Series(chi_scores_churn_contract[1])
```

```python
# Apresentar numeros com decimais sem a notaÃ§Ã£o cientifica
pd.set_option('display.float_format', lambda x: '%.15f' % x)
```

```python
df_chi_scores_churn_contract = pd.DataFrame({'Qui2': score_churn_contract, 'p-value': pvalues_churn_contract})
```

```python
# Analisar Scores e P-Value
df_chi_scores_churn_contract
```

Conforme teste estatÃ­stico (Chi-Square), o P-Value <= 0.05, desta forma rejeitamos a hipÃ³tese nula, ou seja, as variÃ¡veis nÃ£o independentes. Pelo Qui2 alto, podemos afirmar que hÃ¡ uma forte correlaÃ§Ã£o.

```python
# Avaliar a correlaÃ§Ã£o entre uma vÃ¡riavel qualitativa e quantitativa 
# Usar o mesmo teste de Chi-Square, mas transformar a variÃ¡vel quantitativas em qualitativa
#HipÃ³tese: Cliente com menos de 6 meses de contrato Ã© mais propenso ao Churn

# Criar uma nova coluna no DataFrame com base numa condiÃ§Ã£o de outra coluna
# Coluna chamada 'TempoMenor6Meses'ArithmeticError

import numpy as np
```

```python
# Criar coluna nova
df_churn['TempoMenor6Meses'] = np.where(df_churn.tenure<6,'Yes', 'No')
```

```python
# Checar o DataFrame
df_churn.head(5)
```

```python
df_crosstab_churn_tenure = pd.crosstab(df_churn.Churn, df_churn.TempoMenor6Meses)
```

```python
df_crosstab_churn_tenure
```

```python
chi_scores_churn_tenure = chi2_contingency(df_crosstab_churn_tenure)
```

```python
score_churn_tenure = pd.Series(chi_scores_churn_tenure[0])
pvalues_churn_tenure = pd.Series(chi_scores_churn_tenure[1])
```

```python
df_chi_scores_churn_tenure = pd.DataFrame({'Qui2': score_churn_tenure, 'p-value': pvalues_churn_tenure})
```

```python
df_chi_scores_churn_tenure
```

Como o pvalue <= 0.05, rejeitamos a H0, ou seja, aas variÃ¡veis nÃ£o sÃ£o indendependentes. E vale mencionar que com base do Qui2 resultante das 2 anÃ¡lises, a correlaÃ§Ã£o entre Churn e Tempod de Contrato < 6 meses Ã© menor que a correlaÃ§Ã£o entre Churn e Tipo de Contrato.

```python
# CorrelaÃ§Ã£o entre 2 varÃ¡veis numÃ©ricas
# Tenure com TotalCharges
# A intuiÃ§Ã£o Ã© que quanto mais tempo de contrato maior o valor pago.

# CorrelaÃ§Ã£o entre 2 variÃ¡veis numÃ©ricas - Pearson

df_churn.tenure.corr(df_churn.TotalCharges)
```

```python
# MÃ©todo de Tukey - IQR ( DistribuiÃ§Ãµes nÃ£o normal )
# IRQ - Range Interquartil
# IQR = 3 quartil - 1 quartil
q1_TotalCharges_month = df_churn_month.TotalCharges.quantile(0.25)
q3_TotalCharges_month = df_churn_month.TotalCharges.quantile(0.75)
iqr_TotalCharges_month = q3_TotalCharges_month - q1_TotalCharges_month
iqr_TotalCharges_month
```

```python
# Limites inferior e superior
limInferior_TotalCharges_month = q1_TotalCharges_month - (iqr_TotalCharges_month * 1.5)
limSuperiorInf_TotalCharges_month = q3_TotalCharges_month + (iqr_TotalCharges_month * 1.5)
limInferior_TotalCharges_month, limSuperiorInf_TotalCharges_month
```

```python
df_churn_month[(df_churn_month.TotalCharges < limInferior_TotalCharges_month) | (df_churn_month.TotalCharges > limSuperiorInf_TotalCharges_month)]
```
